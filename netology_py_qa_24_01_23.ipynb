```python
import pandas as pd

def get_netology_posts():
    url = 'https://netology.ru/blog/'
    soup = BeautifulSoup(requests.get(url).text)

    news = soup.find_all('div', 'posts__item')

    rows = []

    for el in news:
        date = el.find('div', 'posts__date').text
        # print(date)
        title = el.find('a', 'posts__link').text
        # print(title)
        link = el.find('a', 'posts__link').get('href')
        # print(link)
        category = el.find('span').text
        # print(category)
        rows.append({
            'date': date,
            'title': title,
            'link': link,
            'category': category
        })

    netology = pd.DataFrame(rows)
    return netology


get_netology_posts().to_csv('netology.csv', index=False)
```
